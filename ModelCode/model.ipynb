```python
from keras.layers import Dense, Flatten, Activation, MaxPool2D, Conv2D
from keras.models import Sequential
import numpy as np
import pickle

X = np.array(pickle.load(open("ModelCode/X.pickle", "rb")))
y = np.array(pickle.load(open("ModelCode/y.pickle", "rb")))

X = X/255.0

model = Sequential()

model.add(Conv2D(64, (3,3), input_shape=X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPool2D(3,3))

model.add(Conv2D(128, (3,3)))
model.add(Activation("relu"))
model.add(MaxPool2D(3,3))

model.add(Flatten())
model.add(Dense(128))
model.add(Activation("relu"))

model.add(Dense(12))
model.add(Activation("softmax"))

model.compile(loss="sparse_categorical_crossentropy",
            optimizer="adam",
            metrics=["accuracy"])

model.fit(X, y, batch_size=128, epochs=100, verbose=1, validation_split=0.3)
```

```python
model.save("GarbageClassifier.model")
```
